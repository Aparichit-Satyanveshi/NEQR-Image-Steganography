```python
import numpy as np

def reconstruct_grayscale_from_extraction(
    img_gray_6_extracted,  # (N,N,6)
    N,                     # cover size
    Hs, Ws                 # original secret size
):
    bitstream = []
    for y in range(N):
        for x in range(N):
            bitstream.extend(img_gray_6_extracted[y][x])

    total_secret_bits = 8 * Hs * Ws
    bitstream = bitstream[:total_secret_bits]

    pixels = []
    for i in range(0, len(bitstream), 8):
        byte = bitstream[i:i+8]
        pixels.append(int("".join(map(str, byte)), 2))

    return np.array(pixels, dtype=np.uint8).reshape(Hs, Ws)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def display_secret_reconstruction(secret_gray, title="Secret Image (Reconstructed)", save_path=None):
    img = np.array(secret_gray, dtype=np.uint8)

    plt.figure(figsize=(4, 4))
    plt.imshow(img, cmap="gray", vmin=0, vmax=255)
    plt.axis("off")
    plt.title(title)

    if save_path is not None:
        plt.savefig(save_path, bbox_inches="tight", dpi=300)

    plt.show()
```
